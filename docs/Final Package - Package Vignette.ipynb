```python
###using df from api function to further look into the data
df1= city_ticketmaster.apiget(city = 'San Francisco', password = key)
df1.head()
```

```python
df2 = df1['sales']
dictionary = {}
for rows in df2:
    for key, value in rows.items():
        if key in dictionary.keys():
            dictionary[key].append(value)
        else:
            dictionary[key]=[value]
```
